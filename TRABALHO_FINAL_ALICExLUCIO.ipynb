##**1 IMPORTAÇÃO DOS DADOS LIDOS DO MOTOR**

Como fazer a importação dos dados lidos no físico para o algoritmo no python.



```
# Importa dados
import numpy as np, scipy as sp, matplotlib.pyplot as plt
from scipy import signal
import pandas as pd, numpy as np
URL_DADOS = r'https://github.com/AliceMarquesIFMG/Trabalho-ProcSin/blob/main/MOTOR_BOM_1.csv'
dados = pd.read_csv(URL_DADOS)

# LINHA DO TEMPO
Millis = dados['Millis'].to_numpy()

# MEDIÇÕES
Leituras = dados['Leituras'].to_numpy()

plt.figure(figsize=(10,2))
plt.title("GRÁFICO 1",color = "r",fontsize="18")
plt.xlabel("$Tempo$ [ms]",color = "r",fontsize="14")
plt.ylabel("$Vibração$ [unidade]",color = "r",fontsize="14")
plt.plot(Millis,Leituras,label="Sinal",color = "g")
plt.legend(fontsize="18")
plt.xlim(100,1000)
plt.show()

### SEU CÓDIGO COMEÇA AQUI ###
from scipy import signal
from scipy.signal.fir_filter_design import fft

def fourier(x, t, Hz=False, dB=False, deg=False, Nfft=None):   

     N = len(x)
     T = t[1] - t[0]
    
     if Nfft==None:     
       Nfft = N    
       
     X_f = fft(x, n=Nfft) 
     X_f = X_f/N 
     X_f = X_f[:Nfft//2] 
     X_f[1:] = 2*X_f[1:] 
 
     if Hz==True:
      f0 = 1/(Nfft*T) 
      fHz = np.arange(Nfft)*f0 
      f = fHz[:Nfft//2]     
     else:
       w0 = 2*np.pi/(Nfft*T) 
       wRad = np.arange(Nfft)*w0 
       f = wRad[:Nfft//2]
    
     if dB==False:
      mag = np.abs(X_f)
    
     else:
      mag = 20*np.log10(np.abs(X_f))

     if deg==False:
      pha = np.angle(X_f, deg=0)
     else:
      pha = np.angle(X_f, deg=1)

     return mag, pha, f 

### FIM DO CÓDIGO ###

Leituras = Leituras.reshape(-1)

mag, pha, f = fourier(Leituras, Millis, Hz=True, dB=False, deg=False, Nfft=None)

plt.figure(figsize=(10,2)) 
plt.plot(f, mag) 
plt.title("FFT (módulo), Vibração") 
plt.xlabel("Período (rad/s)") 
plt.ylabel("Amplitude")
#plt.ylim(0,20) 
plt.grid(1); 
```
